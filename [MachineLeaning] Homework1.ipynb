```python
%matplotlib inline
import numpy as np
import cvxpy as cvx
import matplotlib.pyplot as plt
```

```python
m=100
n=10

## didn't make any constraints
A = np.random.rand(m, n)
b = np.random.rand(m)

x = cvx.Variable(n)
#objective = cvx.Minimize(cvx.square(norm2(A*x-b)))
objective = cvx.Minimize(cvx.sum_squares(A*x-b))

prob = cvx.Problem(objective)
result = prob.solve()
```

```python
print('x : ')
print(x.value)
print('\nresult :', result)
```

### Problem 6

```python
n=200
orig = np.zeros(n)
orig[0:100] = 1
noise = np.random.normal(0, 0.1, n)
x_corrupt = orig + noise

plt.plot(x_corrupt)
plt.show()
```

```python
beta=0.5
while beta<2:
    x_con = cvx.Variable(n)
    objective = cvx.Minimize(cvx.norm(x_con - x_corrupt, 2))
    constraints = [cvx.norm(x_con[1:n] - x_con[0:n-1], 1) <= beta]
    prob = cvx.Problem(objective, constraints)
    result = prob.solve()
    
    print('beta : ', beta)
    plt.plot(x_con.value)
    plt.show()
    
    beta+=0.1
```
